```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path= r"C:\Users\Besar\Documents\R\vgames.csv"

data = pd.read_csv(path)
df = pd.DataFrame(data)

df
```

```python
#Check mean Global Sales by each Platform
df1 = df.groupby('Platform', as_index=False)['Global_Sales'].mean()

df1.head()
```

```python
#Check mean Global Sales by each genre
df2 = df.groupby('Genre', as_index=False)['Global_Sales'].mean()

df2.head()
```

```python
#Clean NaN Data
df3 = df.dropna()
df3
```

```python
#Plot to see whether there is a correlation between critic score and critic count
df4 = df3[['Critic_Score','Critic_Count']]
df4.plot.scatter(x='Critic_Score',y='Critic_Count')
df4.head()
```

```python
df4.corr()
```

```python
df4.describe()
```

```python
df4.hist(grid = False)
```

```python
#Plot to see whether there is a correlation between user score and user count
df5 = df3[['User_Score','User_Count']]

#define the x & y data
x = df5['User_Score']
y = df5['User_Count']

#create the scatter plot
plt.plot(x, y, 'o', color = 'cadetblue')

#make sure it is formatted
plt.title('Score vs. Count')
plt.xlabel('User Score')
plt.ylabel('User Count')
plt.show()


```
